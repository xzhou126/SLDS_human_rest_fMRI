```python
from nilearn import image, plotting
import matplotlib.pyplot as plt
import nibabel as nib
```

### Load the overlay and underlay

```python
# overlay = image.load_img('maps/late_map_subcortex_thresholded/late_map_subcortex_thresholded_state6.nii.gz')
# overlay = image.load_img('maps/attractor_map/state1-subcortex.nii.gz')
overlay = image.load_img('maps/region_importance/importance_map_subcortex_thresholded/state1.nii.gz')
underlay = image.load_img('MNI152_2009_template.nii.gz')
```

### Plotting parameters

```python
cut_coords = [5]
cmap = 'RdBu_r'
threshold = 1e-10
vmax = 0.05
vmin = -0.05
```

```python
colorbar = True 
cbar_tick_format = "%.2f" 
dim = 0 
draw_cross = False 
annotate = False 
black_bg = False 
```

### Final image parameters


```python
dpi = 300
width = 10
height = 4
```

subcortical maps, single slice

```python
K = 6
for k in range(K):
    overlay = image.load_img('maps/late_map_subcortex_thresholded/late_map_subcortex_thresholded_state%s.nii.gz'%(k+1))

    display_mode = "y"
    cut_coords = [-2]

    fig,ax = plt.subplots(1,1)
    plotting.plot_stat_map(
        overlay, 
        bg_img=underlay, 
        cut_coords=cut_coords, 
        display_mode=display_mode,
        colorbar=False, 
        cbar_tick_format=cbar_tick_format,
        threshold=threshold,
        annotate=annotate, 
        draw_cross=draw_cross, 
        black_bg=False, 
        cmap=cmap, 
        symmetric_cbar='auto', 
        dim=dim,
        vmax=vmax,
        vmin=vmin,
        figure=fig,
        axes=ax,
    )
    fig.set_size_inches(width, height) 
    fig.savefig("maps/late_map_subcortex_amydala_view/y%s.jpg"%(k+1),dpi=dpi,transparent=True) 
    plt.close()
```

attractor maps

```python
modes = ['y', 'z']
coordinates = [-12, 5]
for k in range(K):
    overlay = image.load_img('maps/attractor_map/state%s-subcortex.nii.gz'%(k+1))

    for i in range(len(modes)):
        display_mode = modes[i]
        cut_coords = [coordinates[i]]

        fig,ax = plt.subplots(1,1)
        plotting.plot_stat_map(
            overlay, 
            bg_img=underlay, 
            cut_coords=cut_coords, 
            display_mode=display_mode,
            colorbar=False, 
            cbar_tick_format=cbar_tick_format,
            threshold=threshold,
            annotate=annotate, 
            draw_cross=draw_cross, 
            black_bg=False, 
            cmap=cmap, 
            symmetric_cbar='auto', 
            dim=dim,
            vmax=0.05,
            vmin=-0.05,
            figure=fig,
            axes=ax,
        )
        fig.set_size_inches(width, height) 
        out_path = "maps/attractor_map/subcortex_hippocampus_view/state%s-%s=%s.jpg"%(k+1,display_mode,coordinates[i])
        fig.savefig(out_path,dpi=dpi,transparent=True) 
        plt.close()
```
